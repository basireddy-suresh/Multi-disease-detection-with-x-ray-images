### Show sample data

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image_dataset_from_directory

# Define the directory containing your training images
training_data_path = 'C:\\Users\\basir\\OneDrive\\Desktop\\Multi-disease detection\\Covid_Detection\\Split_Dataset\\train'

# Load images and labels using image_dataset_from_directory
data_generator = image_dataset_from_directory(
    training_data_path,
    labels='inferred',
    label_mode='categorical',
    batch_size=16,
    image_size=(224, 224)
)

# Define class names
class_names = data_generator.class_names

# Get a batch of images and labels
for images, labels in data_generator.take(1):
    print("Shape of images array:", images.shape)

    # Plot the images
    plt.figure(figsize=(10, 10))
    for i in range(16):  # Adjust the number of images to display as needed
        plt.subplot(4, 4, i + 1)
        image = images[i].numpy().astype("uint8")  # Convert to numpy array and uint8 for plotting
        plt.imshow(image)
        index = np.argmax(labels[i])  # Get the index of the maximum value, which corresponds to the class
        class_name = class_names[index]  # Get the corresponding class name
        plt.title(class_name, color='blue', fontsize=12)
        plt.axis('off')

    plt.show()
```

### Data augmentation and preprocessing on images during the training and validation phases of deep learning model.

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    horizontal_flip=True,
    zoom_range=0.2
)

val_datagen = ImageDataGenerator(rescale=1./255)

training_data = train_datagen.flow_from_directory(
    'C:\\Users\\basir\\OneDrive\\Desktop\\Multi-disease detection\\Covid_Detection\\Split_Dataset\\train',
    target_size=(224, 224),
    class_mode='binary',
    batch_size=32
)

val_data = val_datagen.flow_from_directory(
    'C:\\Users\\basir\\OneDrive\\Desktop\\Multi-disease detection\\Covid_Detection\\Split_Dataset\\val',
    target_size=(224, 224),
    class_mode='binary',
    batch_size=32
)
```

### Model Structure

### MobileNetV2 Model

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Dropout, Flatten, Dense
from keras.optimizers import Adam
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import EarlyStopping, ModelCheckpoint
from PIL import Image

# Define constants
IMAGE_SIZE = (224, 224)
BATCH_SIZE = 64
LEARNING_RATE = 0.001
EPOCHS = 30
```

```python
from keras.applications import MobileNetV2

# Define the MobileNetV2 base model
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze the base model layers
for layer in base_model.layers:
    layer.trainable = False

# Define the top layers for classification
top_layers = Sequential([
    Flatten(input_shape=base_model.output_shape[1:]),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

# Combine the base and top layers
model = Sequential([base_model, top_layers])

# Compile the model
model.compile(optimizer=Adam(lr=LEARNING_RATE), loss='binary_crossentropy', metrics=['accuracy'])

model.summary()


```

### Epochs from training and validation data

```python
# Train the model

history = model.fit(training_data,validation_data=val_data,epochs=EPOCHS, batch_size=64, verbose=1)
```

### Saving the model

```python
from tensorflow.keras.models import save_model

model.save("C:\\Users\\basir\\OneDrive\\Desktop\\Multi-disease detection\\models\\covid_model.h5")
```

### Testing with confusion matrix

```python
import cv2
import os
import seaborn as sns
from tensorflow.keras.models import load_model

TP = 0
TN = 0
FP = 0
FN = 0

model = load_model('C:\\Users\\basir\\OneDrive\\Desktop\\Multi-disease detection\\models\\covid_model.h5')

covid_images_path = 'C:\\Users\\basir\\OneDrive\\Desktop\\Multi-disease detection\\Covid_Detection\\Split_Dataset\\test\\COVID\\'
normal_images_path = 'C:\\Users\\basir\\OneDrive\\Desktop\\Multi-disease detection\\Covid_Detection\\Split_Dataset\\test\\NORMAL\\'

covid_images = os.listdir(covid_images_path)
normal_images = os.listdir(normal_images_path)

for i in covid_images:
    img = cv2.imread(os.path.join(covid_images_path, i))
    if img is None:
        print(f"Failed to read image: {i}")
        continue
    
    img = cv2.resize(img, (224, 224))
    img = img.reshape(1, 224, 224, 3)
    img = img / 255.0
    pred = model.predict(img)
    if pred < 0.5:  
        TP += 1
    else:
        FN += 1

for i in normal_images:
    img = cv2.imread(os.path.join(normal_images_path, i))
    if img is None:
        print(f"Failed to read image: {i}")
        continue
    
    img = cv2.resize(img, (224, 224))
    img = img.reshape(1, 224, 224, 3)
    img = img / 255.0
    pred = model.predict(img)
    if pred > 0.5: 
        TN += 1
    else:
        FP += 1
conf_matrix = [[TP, FP], [FN, TN]]
labels = [['covid', 'normal'], ['covid', 'normal']]

# Plot the heatmap
sns.heatmap(conf_matrix, annot=True, xticklabels=labels[1], yticklabels=labels[0])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

```

### Performance metrics

```python
# Calculate metrics
accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)
sensitivity = TP / (TP + FN)
specificity = TN / (TN + FP)
f1_score = 2 * (precision * recall) / (precision + recall)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("Sensitivity:", sensitivity)
print("Specificity:", specificity)
print("F1 Score:", f1_score)
```

### plotting performance Metrics using matplotlib

```python
import matplotlib.pyplot as plt

# Metrics values
metrics = ['Accuracy', 'Precision', 'Recall', 'Sensitivity', 'Specificity']
values = [accuracy, precision, recall, sensitivity, specificity]

# Convert values to percentages
values_percentage = [val * 100 for val in values]

plt.figure(figsize=(10, 6))
bars = plt.bar(metrics, values_percentage, color=['blue', 'green', 'red', 'orange', 'pink'])
plt.title('Performance Metrics')
plt.xlabel('Metric')
plt.ylabel('Percentage')
plt.ylim(0, 100)  # Set y-axis limit to 0-100 since these are percentages

# Add value labels on top of each bar
for bar, value in zip(bars, values_percentage):
    plt.text(bar.get_x() + bar.get_width() / 2, value, f'{value:.2f}%', ha='center', va='bottom')

plt.show()
```
