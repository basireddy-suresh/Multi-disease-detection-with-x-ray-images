## Show sample data

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image_dataset_from_directory

# Define the directory containing your training images
training_data_path = 'C:\\Users\\basir\\OneDrive\\Desktop\\Multi-disease detection\\Pneumonia Detection\\split_dataset\\train'

# Load images and labels using image_dataset_from_directory
data_generator = image_dataset_from_directory(
    training_data_path,
    labels='inferred',
    label_mode='categorical',
    batch_size=16,
    image_size=(224, 224),
    color_mode='rgb'  # Ensure images are loaded with RGB channels
)

# Define class names
class_names = data_generator.class_names

# Get a batch of images and labels
for images, labels in data_generator.take(1):
    print("Shape of images array:", images.shape)

    # Plot the images
    plt.figure(figsize=(10, 10))
    for i in range(16):  # Adjust the number of images to display as needed
        plt.subplot(4, 4, i + 1)
        image = images[i].numpy().astype("uint8")  # Convert to numpy array and uint8 for plotting
        plt.imshow(image)
        index = np.argmax(labels[i])  # Get the index of the maximum value, which corresponds to the class
        class_name = class_names[index]  # Get the corresponding class name
        plt.title(class_name, color='blue', fontsize=12)
        plt.axis('off')

    plt.show()
```

### Data augmentation and preprocessing on images during the training and validation phases of deep learning model.

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    horizontal_flip=True,
    zoom_range=0.2
)

val_datagen = ImageDataGenerator(rescale=1./255)

training_data = train_datagen.flow_from_directory(
    'C:\\Users\\basir\\OneDrive\\Desktop\\Multi-disease detection\\Pneumonia Detection\\split_dataset\\train',
    target_size=(224, 224),
    class_mode='binary',
    batch_size=32
)

val_data = val_datagen.flow_from_directory(
    'C:\\Users\\basir\\OneDrive\\Desktop\\Multi-disease detection\\Pneumonia Detection\\split_dataset\\val',
    target_size=(224, 224),
    class_mode='binary',
    batch_size=32
)
```

### Model Structure-CNN

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization

# Define parameters
img_size = (224, 224)
channels = 3
class_count = 2  # Assuming binary classification

# Create CNN model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(img_size[0], img_size[1], channels)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

# Compile model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
epochs = 10
history = model.fit(training_data, epochs=epochs, verbose=1, validation_data=val_data)
```

### Evaluate the model performance using matplotlib

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(20, 5))

# Plot Training and Validation Loss
plt.subplot(1, 2, 1)
train_loss = history.history["loss"]
val_loss = history.history["val_loss"]
epochs = range(1, len(train_loss) + 1)
plt.plot(epochs, train_loss, label="Training loss")
plt.plot(epochs, val_loss, label="Validation loss")
plt.title("Training and Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()

# Plot Training and Validation Accuracy
plt.subplot(1, 2, 2)
train_acc = history.history["accuracy"]
val_acc = history.history["val_accuracy"]
plt.plot(epochs, train_acc, label="Training accuracy")
plt.plot(epochs, val_acc, label="Validation accuracy")
plt.title("Training and Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()

plt.show()
```

### Testing with confusion metrics

```python
import cv2
import os
import seaborn as sns
from tensorflow.keras.models import load_model

TP = 0
TN = 0
FP = 0
FN = 0

model = load_model('C:\\Users\\basir\\OneDrive\\Desktop\\Multi-disease detection\\models\\Pneumonia_model.h5')

normal_images_path = 'C:\\Users\\basir\\OneDrive\\Desktop\\Multi-disease detection\\Pneumonia Detection\\split_dataset\\test\\NORMAL'
pneumonia_images_path = 'C:\\Users\\basir\\OneDrive\\Desktop\\Multi-disease detection\\Pneumonia Detection\\split_dataset\\test\\PNEUMONIA'

normal_images = os.listdir(normal_images_path)
pneumonia_images = os.listdir(pneumonia_images_path)

for i in normal_images:
    img = cv2.imread(os.path.join(normal_images_path, i))
    if img is None:
        print(f"Failed to read image: {i}")
        continue
    
    img = cv2.resize(img, (224, 224))
    img = img.reshape(1, 224, 224, 3)
    img = img / 255.0
    pred = model.predict(img)
    if pred < 0.5:  
        TP += 1
    else:
        FN += 1

for i in pneumonia_images:
    img = cv2.imread(os.path.join(pneumonia_images_path, i))
    if img is None:
        print(f"Failed to read image: {i}")
        continue
    
    img = cv2.resize(img, (224, 224))
    img = img.reshape(1, 224, 224, 3)
    img = img / 255.0
    pred = model.predict(img)
    if pred > 0.5: 
        TN += 1
    else:
        FP += 1
conf_matrix = [[TP, FP], [FN, TN]]
labels = [['pneumonia', 'normal'], ['pneumonia', 'normal']]

# Plot the heatmap
sns.heatmap(conf_matrix, annot=True, xticklabels=labels[1], yticklabels=labels[0])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

```

### Evaluate performance metrics

```python
# Calculate metrics
accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)
sensitivity = TP / (TP + FN)
specificity = TN / (TN + FP)
f1_score = 2 * (precision * recall) / (precision + recall)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("Sensitivity:", sensitivity)
print("Specificity:", specificity)
print("F1 Score:", f1_score)
```

### Plotting performance metrics using matplotlib

```python
import matplotlib.pyplot as plt

# Metrics values
metrics = ['Accuracy', 'Precision', 'Recall', 'Sensitivity', 'Specificity']
values = [accuracy, precision, recall, sensitivity, specificity]

# Convert values to percentages
values_percentage = [val * 100 for val in values]

plt.figure(figsize=(10, 6))
bars = plt.bar(metrics, values_percentage, color=['blue', 'green', 'red', 'orange', 'pink'])
plt.title('Performance Metrics')
plt.xlabel('Metric')
plt.ylabel('Percentage')
plt.ylim(0, 100)  # Set y-axis limit to 0-100 since these are percentages

# Add value labels on top of each bar
for bar, value in zip(bars, values_percentage):
    plt.text(bar.get_x() + bar.get_width() / 2, value, f'{value:.2f}%', ha='center', va='bottom')

plt.show()
```
