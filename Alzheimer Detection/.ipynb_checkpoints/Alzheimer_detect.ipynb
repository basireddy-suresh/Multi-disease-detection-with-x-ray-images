### Distribution of Alzheimer dataset with their classes in Bar chart using Matplotlib

```python
import matplotlib.pyplot as plt

# Define the dataset directories
dataset_dirs = [
    "C:\\Users\\basir\\OneDrive\\Desktop\\Multi-disease detection\\Alzheimer Detection\\Alzheimer_Dataset"
]

# Initialize counts for each class across all datasets
class_counts = {class_name: 0 for class_name in class_names}

# Iterate through each dataset directory
for dataset_dir in dataset_dirs:
    # Create a data generator for the current dataset directory
    data_generator = tf.keras.preprocessing.image_dataset_from_directory(
        dataset_dir,
        seed=123,
        image_size=(IMG_HEIGHT, IMG_WIDTH),
        batch_size=64,
        label_mode='int'
    )
    
    # Iterate through the dataset and count occurrences of each class
    for images, labels in data_generator:
        for label in labels.numpy():
            class_name = class_names[label]
            class_counts[class_name] += 1

# Extract class names and counts
class_names = list(class_counts.keys())
counts = list(class_counts.values())

# Define colors for each class
colors = ['blue', 'green', 'red', 'orange']  # Add more colors if needed

# Plot the bar chart with different colors for each class
plt.figure(figsize=(10, 6))
bars = plt.bar(class_names, counts, color=colors)
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Dataset Distribution')

# Add count labels on top of each bar
for bar, count in zip(bars, counts):
    plt.text(bar.get_x() + bar.get_width() / 2, count, str(count), ha='center', va='bottom')

plt.show()
```

### Model Structure

### CNN Model


```python
model = keras.models.Sequential()
model.add(keras.layers.experimental.preprocessing.Rescaling(1./255, input_shape=(IMG_HEIGHT,IMG_WIDTH, 3)))
model.add(keras.layers.Conv2D(filters=16,kernel_size=(3,3),padding='same',activation='relu',kernel_initializer="he_normal"))
model.add(keras.layers.MaxPooling2D(pool_size=(2,2)))


model.add(keras.layers.Conv2D(filters=32,kernel_size=(3,3),padding='same',activation='relu',kernel_initializer="he_normal"))
model.add(keras.layers.MaxPooling2D(pool_size=(2,2)))

model.add(keras.layers.Dropout(0.20))

model.add(keras.layers.Conv2D(filters=64,kernel_size=(3,3),padding='same',activation='relu',kernel_initializer="he_normal"))
model.add(keras.layers.MaxPooling2D(pool_size=(2,2)))

model.add(keras.layers.Dropout(0.25))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(128,activation="relu",kernel_initializer="he_normal"))
model.add(keras.layers.Dense(64,"relu"))
model.add(keras.layers.Dense(4,"softmax"))
```

```python
model.compile(loss="sparse_categorical_crossentropy",
              optimizer = "Adam",metrics=["accuracy"])
```

```python
model.summary()
```

```python
hist = model.fit(train_ds,validation_data=val_ds,epochs=30, batch_size=64, verbose=1)
```

### Plotting Training and validation based on epochs

```python
get_ac = hist.history['accuracy']
get_los = hist.history['loss']
val_acc = hist.history['val_accuracy']
val_loss = hist.history['val_loss']
```

```python
epochs = range(len(get_ac))
plt.plot(epochs, get_ac, 'g', label='Accuracy of Training Data')
plt.plot(epochs, val_acc, 'r', label='Accuracy of Validation Data')
plt.title('Training and Validation Accuracy')
plt.legend(loc=0)
plt.figure()

plt.plot(epochs, get_los, 'g', label='Loss of Training Data')
plt.plot(epochs, val_loss, 'r', label='Loss of Validation Data')
plt.title('Training and Validation Loss')
plt.legend(loc=0)
plt.figure()
plt.show()
```

### Confusion Matrix

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Initialize empty lists to store true labels and predicted labels
true_labels = []
predicted_labels = []

# Iterate through the test dataset and make predictions
for images, labels in test_ds:
    # Get the predicted probabilities for the images
    predictions = model.predict(images)
    # Convert predicted probabilities to class labels
    predicted_classes = np.argmax(predictions, axis=1)
    # Append true and predicted labels to the respective lists
    true_labels.extend(labels.numpy())
    predicted_labels.extend(predicted_classes)

# Calculate the confusion matrix
conf_matrix = confusion_matrix(true_labels, predicted_labels)

# Get class names
class_names = test_ds.class_names

# Plot the confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
TP = np.diag(conf_matrix)
FP = np.sum(conf_matrix, axis=0) - TP
FN = np.sum(conf_matrix, axis=1) - TP
TN = np.sum(conf_matrix) - (TP + FP + FN)

accuracy = np.sum(TP) / np.sum(conf_matrix)
precision = np.mean(TP / (TP + FP))
recall = np.mean(TP / (TP + FN))
sensitivity = np.mean(TP / (TP + FN))
specificity = np.mean(TN / (TN + FP))
f1_score = 2 * (precision * recall) / (precision + recall)

# Print metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("Sensitivity:", sensitivity)
print("Specificity:", specificity)
print("F1 Score:", f1_score)
```

```python
import matplotlib.pyplot as plt

# Metrics values
metrics = ['Accuracy', 'Precision', 'Recall', 'Sensitivity', 'Specificity']
values = [accuracy, precision, recall, sensitivity, specificity]

# Convert values to percentages
values_percentage = [val * 100 for val in values]

plt.figure(figsize=(10, 6))
bars = plt.bar(metrics, values_percentage, color=['blue', 'green', 'red', 'orange', 'pink'])
plt.title('Performance Metrics')
plt.xlabel('Metric')
plt.ylabel('Percentage')
plt.ylim(0, 100)  # Set y-axis limit to 0-100 since these are percentages

# Add value labels inside each bar
for bar, value in zip(bars, values_percentage):
    plt.text(bar.get_x() + bar.get_width() / 2, value / 2, f'{value:.2f}%', ha='center', va='center', color='black')

plt.show()
```
